```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, TrainingArguments, Trainer
import torch
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFacePipeline
```

```python
!pip install datasets
```

```python
model_id = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model_id)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token  

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_4bit=True,  
    torch_dtype=torch.float16  
)
```

```python
from datasets import load_dataset
dataset = load_dataset("json", data_files="/content/Dataset_chatbot.jsonl")
```

```python
prompt_template = """
You are a Verizon customer support chatbot. Your goal is to assist customers with their queries related to plans, usage, orders, perks, and troubleshooting. Maintain context throughout the conversation and provide accurate responses based on the following dataset.

### Conversation History:
{history}

### Customer Query:
{input}

### Response:
{response}
"""
```

```python
def preprocess_function(examples):
    prompts = [
        prompt_template.format(
            history="",  
            input=instruction,
            response=response
        )
        for instruction, response in zip(examples["instruction"], examples["response"])
    ]

    tokenized_outputs = tokenizer(
        prompts,
        truncation=True,
        padding="max_length",
        max_length=512,
        return_tensors="pt"
    )
    tokenized_outputs["labels"] = tokenized_outputs["input_ids"].clone()

    return tokenized_outputs

tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
split_dataset = tokenized_dataset["train"].train_test_split(test_size=0.1)  
train_dataset = split_dataset["train"]
eval_dataset = split_dataset["test"]
```

```python
from peft import LoraConfig, get_peft_model
```

```python
lora_config = LoraConfig(
    r=8,  
    lora_alpha=32,  
    target_modules=["q_proj", "v_proj"], 
    lora_dropout=0.1,  
    bias="none", 
    task_type="CAUSAL_LM"  
)

model = get_peft_model(model, lora_config)
for name, param in model.named_parameters():
    if "lora" in name:
        print(f"{name}: requires_grad={param.requires_grad}")
```

```python
model.gradient_checkpointing_enable(gradient_checkpointing_kwargs={"use_reentrant": False})
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=4,  
    num_train_epochs=3,
    save_steps=10_000,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=500,
    evaluation_strategy="steps",
    eval_steps=500,              
    learning_rate=5e-5,
    fp16=False,  
    gradient_checkpointing=False,  
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```

```python
model_id = "meta-llama/Llama-2-7b-chat-hf"  
```

```python
trainer.save_model("./fine-tuned-verizon-chatbot")
tokenizer.save_pretrained("./fine-tuned-verizon-chatbot")
```

```python
fine_tuned_model = AutoModelForCausalLM.from_pretrained(
    "./fine-tuned-verizon-chatbot",
    device_map="auto",  
    load_in_4bit=True,  
    torch_dtype=torch.float16  
    
)
fine_tuned_tokenizer = AutoTokenizer.from_pretrained("./fine-tuned-verizon-chatbot")
memory = ConversationBufferMemory(max_history_size=5)  
```

```python
def verizon_chatbot(query):
  
    history = memory.load_memory_variables({})["history"]
    prompt = prompt_template.format(
        history=history,
        input=query,
        response=""
    )
    inputs = fine_tuned_tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = fine_tuned_model.generate(**inputs, max_length=512)
    response = fine_tuned_tokenizer.decode(outputs[0], skip_special_tokens=True)
    memory.save_context({"input": query}, {"output": response})

    return response
```

```python
if __name__ == "__main__":
    print("Welcome to the Verizon Chatbot! Type 'exit' to end the conversation.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            break
        response = verizon_chatbot(user_input)
        print(f"Verizon Chatbot: {response}")
```
